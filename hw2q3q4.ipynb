```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
clf = DecisionTreeClassifier(random_state=0,criterion='entropy')
```

```python
D3 = pd.read_table('Dbig.txt', names=['x1','x2','y'], sep=' ')
D3_8192, D3_test = train_test_split(D3, test_size=1 - 0.8192, random_state=0)
D3_2048 = D3_8192.sample(2048,random_state=0)
D3_512  = D3_8192.sample(512,random_state=0)
D3_128  = D3_8192.sample(128,random_state=0)
D3_32   = D3_8192.sample(32,random_state=0)
```

```python
clf5 = clf.fit(D3_8192[['x1','x2']], D3_8192['y'])
D3_test['pred_8192'] = clf.predict(D3_test[['x1','x2']])

clf4 = clf.fit(D3_2048[['x1','x2']], D3_2048['y'])
D3_test['pred_2048'] = clf.predict(D3_test[['x1','x2']])

clf3 = clf.fit(D3_512[['x1','x2']], D3_512['y'])
D3_test['pred_512'] = clf.predict(D3_test[['x1','x2']])

clf2 = clf.fit(D3_128[['x1','x2']], D3_128['y'])
D3_test['pred_128'] = clf.predict(D3_test[['x1','x2']])

clf1 = clf.fit(D3_32[['x1','x2']], D3_32['y'])
D3_test['pred_32'] = clf.predict(D3_test[['x1','x2']])
```
